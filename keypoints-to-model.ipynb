```python
from sklearn.model_selection import train_test_split

training_data, testing_data = train_test_split(df, test_size=0.2, random_state=25)

print(f"No. of training examples: {training_data.shape[0]}")
print(f"No. of testing examples: {testing_data.shape[0]}")

training_data = training_data.copy()
testing_data = testing_data.copy()
```

```python
training_data.dtypes, testing_data.dtypes
```

```python
len(training_data['selected_frames'][0])
```

```python
#shape=(1, 40, 2, 21, 2), 1 vdo, 40 frames, 2 hands, 21 keypoints of 2 dimensions. 21*2 = 42 for a hand. 2 hands so, 42*2 = 84
X_train = pd.DataFrame(columns = [i for i in range(84*40)])
X_train 
X_test = pd.DataFrame(columns = [i for i in range(84*40)])
X_test
```

```python
# row=0
# for frames in training_data['selected_frames']:
#     print(len(frames), row)
#     row = row+1
```

```python
for frames in training_data['selected_frames']:
    row=[None]*3360
    i=0
    for hands in frames:
        for keypoints in hands:
            for keypoint in keypoints:
                for cordinate in keypoint:
                    row[i]=cordinate
                    i = i+1
    X_train.loc[len(X_train.index)]=row
```

```python
for frames in testing_data['selected_frames']:
    row=[None]*3360
    i=0
    for hands in frames:
        for keypoints in hands:
            for keypoint in keypoints:
                for cordinate in keypoint:
                    row[i]=cordinate
                    i = i+1
    X_test.loc[len(X_test.index)]=row
```
